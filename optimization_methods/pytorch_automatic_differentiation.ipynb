Поиск производной функции одной переменной

```python
import torch

# ШАГ 1: Создаем переменную x, равную 4.0
# Обязательно делаем число вещественным (с точкой), так как градиенты не считаются для целых чисел.
# Флаг requires_grad указывает PyTorch, что для x нужно собирать граф вычислений.
x = torch.tensor(4.0, requires_grad=True)

# ШАГ 2: Задаем саму функцию (Прямой проход)
# PyTorch автоматически построит вычислительный граф для переменной y.
y = x**2 + 3 * x + 1

# ШАГ 3: Вычисляем производную dy/dx с помощью модуля autograd
# В функцию grad нужно передать: Что дифференцируем (outputs) и По чему дифференцируем (inputs)
# Внимание: inputs ожидает список или кортеж, поэтому пишем [x]
dy_dx = torch.autograd.grad(outputs=y, inputs=[x])[0]

# ШАГ 4: Проверяем результат
print(f"Значение функции в точке x=4: {y.item()}")  # Должно быть 4^2 + 3*4 + 1 = 29
print(f"Вычисленная производная dy/dx: {dy_dx.item()}")  # Должно быть 11

# --- ТВОЕ ЗАДАНИЕ ДЛЯ ЗАКРЕПЛЕНИЯ ---
# Измени код ниже, чтобы посчитать производную для функции f(x) = ln(x) + sin(x) в точке x = 1.0
# Подсказка: используй torch.log(x) для натурального логарифма и torch.sin(x)

x_new = torch.tensor(1.0, requires_grad=True)

# 1. Напиши формулу для y_new: ln(x) + sin(x)
y_new = torch.log(x_new) + torch.sin(x_new)  # ТВОЙ КОД ЗДЕСЬ

# 2. Посчитай производную dy_new_dx_new с помощью torch.autograd.grad
dy_dx_new = torch.autograd.grad(outputs=y_new, inputs=[x_new])[0]  # ТВОЙ КОД ЗДЕСЬ
# 3. Распечатай результат через print()
print(f"derivative new func: {dy_dx_new.item()}")
```

Частные производные и Градиент функции многих переменных

```python
# ШАГ 1: Создаем вектор параметров w = [x, y] со значениями [2.0, 0.0]
w = torch.tensor([2.0, 0.0], requires_grad=True)
print(w)
# Чтобы нам было удобно писать формулу, «вытащим» x и y из вектора w:
x = w[0]
y = w[1]

# ШАГ 2: Задаем функцию (Прямой проход)
# Подсказка: для экспоненты e^y используйте torch.exp(y)
f = x**2 * y + torch.exp(y)

# ШАГ 3: Вычисляем градиент df/dw
# Так как w — это вектор, PyTorch посчитает производную f по каждому элементу w
grad_f = torch.autograd.grad(outputs=f, inputs=[w])[0]

# ШАГ 4: Выводим результат
print(f"Значение функции: {f.item()}")  # Должно быть 2^2 * 0 + e^0 = 1.0
print(f"Градиент [df/dx, df/dy]: {grad_f.tolist()}")  # Должно быть [0.0, 5.0]


# --- ТВОЕ ЗАДАНИЕ ДЛЯ ЗАКРЕПЛЕНИЯ ---
# Посчитай градиент функции Розенброка (это классическая функция-банан из курса Меркулова)
# Формула для двух переменных: f(x, y) = 100 * (y - x^2)^2 + (1 - x)^2
# Найди градиент в точке (x = 0.0, y = 0.0)

w_rosen = torch.tensor([0.0, 0.0], requires_grad=True)

x_r = w_rosen[0]  # 1. Вытащи x_r и y_r из w_rosen
y_r = w_rosen[1]
f_rosen = 100 * (y_r - x_r**2) ** 2 + (1 - x_r) ** 2  # 3. Посчитай градиент grad_rosen с помощью torch.autograd.grad
grad_rosen = torch.autograd.grad(outputs=f_rosen, inputs=[w_rosen])  # 2. Напиши формулу f_rosen
grad_rosen_ans = grad_rosen[0]
print(grad_rosen_ans.tolist())  # 4. Выведи результат на экран
```

Векторные операции и Якобиан (Шаг к продвинутому автограду)

```python
# 1. Задаем фиксированную матрицу A размера 2x2
A = torch.tensor([[3.0, 5.0], [1.0, 2.0]])

# 2. Задаем вектор переменных x = [x1, x2]
x = torch.tensor([1.0, 1.0], requires_grad=True)

# 3. Прямой проход: матричное умножение A на x
# В PyTorch матричное умножение делается через оператор @
y = A @ x

# СЕЙЧАС y — это вектор [8.0, 3.0].
# Если мы просто вызовем torch.autograd.grad(y, x), PyTorch выдаст ошибку!
# Чтобы посчитать градиент, нам нужно передать вектор весов (grad_outputs),
# который «спроецирует» вектор y в скаляр.
# Передадим вектор [1.0, 0.0] — это значит, что мы хотим найти производную ТОЛЬКО первой компоненты y по x.

dy2_dx = torch.autograd.grad(outputs=y, inputs=[x], grad_outputs=torch.tensor([0.0, 1.0]))

print(f"Производная второй строчки по x: {dy2_dx[0].tolist()}")
# Подсказка: она должна совпасть с ПЕРВОЙ строчкой матрицы A, то есть [3.0, 5.0]


# --- ТВОЕ ЗАДАНИЕ ДЛЯ ЗАКРЕПЛЕНИЯ ---
# Измени код выше (или напиши ниже), чтобы найти производную ВТОРОЙ компоненты y по x.
# Какой `grad_outputs` нужно передать в функцию?
# С какой строчкой матрицы A должен совпасть ответ?
```

```python
import torch

A = torch.tensor([[3.0, 5.0], [1.0, 2.0]])
x = torch.tensor([1.0, 1.0], requires_grad=True)
y = A @ x

# ШАГ 1: Считаем производную ПЕРВОЙ строчки
# Обязательно добавляем retain_graph=True, чтобы граф НЕ стирался!
dy1_dx = torch.autograd.grad(outputs=y, inputs=[x], grad_outputs=torch.tensor([1.0, 0.0]), retain_graph=True)
print(f"Производная первой строчки по x: {dy1_dx[0].tolist()}")

# ШАГ 2: Теперь мы можем спокойно считать производную ВТОРОЙ строчки
# Здесь граф уже можно не сохранять (если дальше мы не планируем брать третью производную)
dy2_dx = torch.autograd.grad(outputs=y, inputs=[x], grad_outputs=torch.tensor([0.0, 1.0]))
print(f"Производная второй строчки по x: {dy2_dx[0].tolist()}")
```

 Главная «фишка» лекции — Hessian-Vector Product (HVP)

```python
import torch

# Задаем симметричную матрицу матрицы вторых производных
A = torch.tensor([[4.0, 1.0], [1.0, 3.0]])

x = torch.tensor([1.0, 2.0], requires_grad=True)
v = torch.tensor([0.5, 1.5])  # Вектор направления

# Наша функция — квадратичная форма. Её Гессиан в любой точке равен матрице A.
# Формула: f = 0.5 * x^T * A * x
f = 0.5 * torch.dot(x, A @ x)

# ШАГ 1: Первый проход автограда. Находим градиент функции f по переменной x.
# ВАЖНО: передаем create_graph=True, чтобы PyTorch построил вычислительный граф
# для самого градиента (иначе мы не сможем взять вторую производную!).
grad_f = torch.autograd.grad(outputs=f, inputs=[x], create_graph=True)[0]

# ШАГ 2: Считаем скалярное произведение grad_f и вектора v
# ТВОЙ КОД ЗДЕСЬ: используй torch.dot(grad_f, v)
scalar_product = torch.dot(grad_f, v)

# ШАГ 3: Второй проход автограда. Находим градиент от scalar_product по переменной x.
# ТВОЙ КОД ЗДЕСЬ: используй torch.autograd.grad
hvp = torch.autograd.grad(outputs=scalar_product, inputs=[x])

# ШАГ 4: Проверка. Аналитический ответ равен просто A @ v
analytical_ans = A @ v

# Раскомментируй строчку ниже, когда допишешь код:
print(f"Результат HVP через автоград: {hvp[0].tolist()}")
print(f"Математический ответ (A @ v): {analytical_ans.tolist()}")
```
